```python
import os
import sys
import random
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

sys.path.append(os.path.abspath(".."))

from app.vision.inference import load_model, predict

translate = {
    "cane": "dog", "cavallo": "horse", "elefante": "elephant", "farfalla": "butterfly", 
    "gallina": "chicken", "gatto": "cat", "mucca": "cow", "pecora": "sheep", 
    "ragno": "spider", "scoiattolo": "squirrel",
    "dog": "cane", "horse": "cavallo", "elephant": "elefante", "butterfly": "farfalla", 
    "chicken": "gallina", "cat": "gatto", "cow": "mucca", "spider": "ragno", "squirrel": "scoiattolo"
}

# 
model_path = "../mnt/data/models/vision_model.h5"
data_path = "../mnt/data/check/"

#  
print("Loading model...")
model = load_model(model_path)

#     
all_images = glob.glob(os.path.join(data_path, "*/*.*"))[:100]

print(f"Found {len(all_images)} images for prediction.")

results = []
correct_predictions = 0

#   100 
for img in all_images:
    class_idx, confidence = predict(model, img)
    class_labels = ['cane', 'cavallo', 'elefante', 'farfalla', 'gallina', 'gatto', 'mucca', 'pecora', 'ragno', 'scoiattolo']
    predicted_class = class_labels[class_idx]
    
    # ,   
    actual_class = os.path.basename(img).split('_')[0]  # ,  '    
    if translate.get(predicted_class) == actual_class or predicted_class == actual_class:
        correct_predictions += 1
    
    results.append((img, translate[predicted_class], confidence))

#  
accuracy = correct_predictions / len(all_images) * 100
print(f"\nAccuracy on 100 images: {accuracy:.2f}%")

#  10  
print("\nRandom 10 predictions:")
sample = random.sample(results, 10)
for img, label, conf in sample:
    print(f"Image: {os.path.basename(img)} -> Predicted: {label} (Confidence: {conf:.2f})")
    
    #     
    img_data = mpimg.imread(img)
    plt.figure(figsize=(5, 5))
    plt.imshow(img_data)
    plt.title(f"Predicted: {label} (Confidence: {conf:.2f})")
    plt.axis('off')
    plt.show()

print(" Prediction completed.")
```

```python
import os
import sys

sys.path.append(os.path.abspath(".."))


from app.ner.inference import NERInference
from app.ner.model import NERModel
from tensorflow.keras.preprocessing.text import Tokenizer


texts = ["There is a cow in the picture", "A cat is jumping", "I see a horse"]
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)

model = NERModel(vocab_size=10000, max_len=100, num_labels=10).model
inference = NERInference(tokenizer, model)

for text in texts:
    prediction = inference.predict(text)
    print(f"Input: {text}")
    print(f"Prediction: {prediction}")
```

```python
import numpy as np

translate = {"cane": "dog", "cavallo": "horse", "elefante": "elephant", 
             "farfalla": "butterfly", "gallina": "chicken", "gatto": "cat", 
             "mucca": "cow", "pecora": "sheep", "scoiattolo": "squirrel", 
             "dog": "cane", "horse": "cavallo", "elephant": "elefante", 
             "butterfly": "farfalla", "chicken": "gallina", "cat": "gatto", 
             "cow": "mucca", "spider": "ragno", "squirrel": "scoiattolo"}

classes = list(translate.keys())[:10]  #  

def interpret_predictions(predictions):
    result = []
    for token_preds in predictions[0]:
        class_idx = np.argmax(token_preds)  #     
        result.append(classes[class_idx])
    return result

for text in texts:
    prediction = inference.predict(text)
    labels = interpret_predictions(prediction)
    print(f"Input: {text}")
    print(f"NER Labels: {labels}")
```
